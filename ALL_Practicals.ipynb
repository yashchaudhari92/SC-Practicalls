```python
import numpy as np

# Set seed for reproducible results
np.random.seed(42)

# ===============================================
# 1. Data Definition and Normalization
# ===============================================
# Input features X: [Hours_Studied, Pages_Read]
X = np.array([[2, 9], [1, 5], [3, 6]], float)
# Normalize X by dividing by the maximum value in each column (max-min scaling where min=0)
X = X / np.amax(X, axis=0)

# Output target y: [Exam_Score]
y = np.array([[92], [86], [89]], float)
# Normalize y to be between 0 and 1
y = y / 100.0

# ===============================================
# 2. Activation Functions
# ===============================================
# Sigmoid activation function
sig = lambda x: 1 / (1 + np.exp(-x))

# Derivative of the Sigmoid function (for backpropagation)
# Note: It takes the *output* of the sigmoid function as input
dsig = lambda x: x * (1 - x)

# ===============================================
# 3. Hyperparameters and Initialization
# ===============================================
lr = 0.1         # Learning Rate
epochs = 7000    # Number of training iterations

# Initialize weights and biases for the hidden layer (2 inputs, 3 neurons)
W1 = np.random.uniform(size=(2, 3))
b1 = np.random.uniform(size=(1, 3))

# Initialize weights and biases for the output layer (3 inputs, 1 output neuron)
W2 = np.random.uniform(size=(3, 1))
b2 = np.random.uniform(size=(1, 1))

# ===============================================
# 4. Training Loop
# ===============================================
for _ in range(epochs):

    # --- A. Forward Propagation ---
    # Hidden Layer Calculation
    # h = activation( (Input * W1) + b1 )
    h = sig(np.dot(X, W1) + b1)

    # Output Layer Calculation
    # out = activation( (Hidden_Output * W2) + b2 )
    out = sig(np.dot(h, W2) + b2)

    # --- B. Backpropagation ---

    # 1. Output Layer Error and Delta
    # Error: Target - Predicted
    err_out = y - out
    # Delta_out = Error * Sigmoid_Derivative_of_Output
    d_out = err_out * dsig(out)

    # 2. Hidden Layer Error and Delta
    # Error: Delta_out dotted with Transposed_W2 (distribute error backwards)
    err_h = d_out.dot(W2.T)
    # Delta_h = Error * Sigmoid_Derivative_of_Hidden
    d_h = err_h * dsig(h)

    # --- C. Update Weights and Biases ---

    # Update Output Layer Weights and Bias
    W2 += h.T.dot(d_out) * lr
    b2 += np.sum(d_out, axis=0, keepdims=True) * lr

    # Update Hidden Layer Weights and Bias
    W1 += X.T.dot(d_h) * lr
    b1 += np.sum(d_h, axis=0, keepdims=True) * lr

# ===============================================
# 5. Results and Evaluation
# ===============================================
print("--- Training Completed ---")
print("\nInput (Normalized):\n", X)
print("\nActual Output (Normalized):\n", y)
print("\nPredicted Output (Normalized, Rounded):\n", np.round(out, 8))

# Accuracy Metric (Mean Absolute Percentage Error based accuracy)
mape = np.mean(np.abs((y - out) / y)) * 100
accuracy = 100 - mape
print("\nAccuracy (100 - MAPE): **{:.2f}%**".format(accuracy))
```

```python
import numpy as np
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error

# Set seed for reproducibility
np.random.seed(0)

# ===============================================
# 1. Data Generation
# ===============================================
# Generate 100 2D random input points (X)
X = np.random.rand(100, 2)
# Generate target output (y) using a non-linear function
y = np.sin(X[:, 0]) + np.cos(X[:, 1])

# ===============================================
# 2. RBF Function Definition
# ===============================================
def rbf(x, c, s):
    """
    Computes the Gaussian Radial Basis Function output.

    Args:
        x (np.array): Input vector.
        c (np.array): Center vector of the RBF neuron.
        s (float): Spread parameter (sigma).
    """
    # Formula: exp( - ||x - c||^2 / (2 * s^2) )
    return np.exp(-np.linalg.norm(x - c) ** 2 / (2 * s ** 2))

# ===============================================
# 3. Center and Spread Calculation
# ===============================================
# Choose the RBF centers using K-Means clustering
K_CENTERS = 10
kmeans = KMeans(n_clusters=K_CENTERS, random_state=0, n_init=10).fit(X)
centers = kmeans.cluster_centers_

# Calculate the spread parameter (sigma) based on the maximum distance
# between chosen centers (heuristic approach)
d_max = np.max(cdist(centers, centers, 'euclidean'))
sigma = d_max / np.sqrt(2 * len(centers))

# ===============================================
# 4. RBF Layer Output (Hidden Layer) Computation
# ===============================================
# R is the matrix where R[i, j] is the output of RBF neuron 'j' for input 'i'
R = np.zeros((X.shape[0], len(centers)))
for i in range(X.shape[0]):
    for j in range(len(centers)):
        R[i, j] = rbf(X[i], centers[j], sigma)

# ===============================================
# 5. Output Weight Calculation (Linear Solver)
# ===============================================
# Compute the output weights (W) using the pseudo-inverse (Least Squares solution)
# W = pinv(R) * y
W = np.dot(np.linalg.pinv(R), y)

# ===============================================
# 6. RBF Network Prediction Function
# ===============================================
def rbf_network(X, centers, sigma, W):
    """
    Performs the full RBF network prediction.
    """
    R = np.zeros((X.shape[0], len(centers)))

    # Compute RBF layer output for the input data X
    for i in range(X.shape[0]):
        for j in range(len(centers)):
            R[i, j] = rbf(X[i], centers[j], sigma)

    # Calculate final output: y_pred = R * W
    return np.dot(R, W)

# ===============================================
# 7. Prediction and Evaluation
# ===============================================
# Make predictions on the training data
y_pred = rbf_network(X, centers, sigma, W)

# Evaluate the model's performance
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.6f}")
```

```python

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from tensorflow.keras.datasets import mnist
import seaborn as sns

# [ 1 ] Load MNIST Dataset
print("Loading MNIST Dataset...")
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")

# [ 2 ] Exploratory Data Analysis (EDA)
# Display some sample digits
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.suptitle("Sample Handwritten Digits from MNIST", fontsize=14)
plt.show()

# Count distribution of labels
unique, counts = np.unique(y_train, return_counts=True)
plt.bar(unique, counts, color='royalblue')
plt.title("Label Distribution in Training Set")
plt.xlabel("Digit")
plt.ylabel("Count")
plt.show()

# Compute average pixel intensity per digit
avg_intensity = [np.mean(X_train[y_train == i]) for i in range(10)]
plt.bar(range(10), avg_intensity, color='orange')
plt.title("Average Pixel Intensity per Digit")
plt.xlabel("Digit")
plt.ylabel("Average Intensity")
plt.show()

print(f"Standard deviation of intensities: {np.std(avg_intensity)}")

# [ 3 ] Data Preprocessing
# Flatten images (28x28 -> 784)
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# Normalize data
X_train_flat = X_train_flat / 255.0
X_test_flat = X_test_flat / 255.0

# For faster testing, we can use smaller subsets (optional)
# X_train_flat, _, y_train, _ = train_test_split(X_train_flat, y_train, train_size=10000, stratify=y_train)
# X_test_flat, _, y_test, _ = train_test_split(X_test_flat, y_test, train_size=2000, stratify=y_test)


# [ 4 ] Train SVM classifier
print("\nTraining Support Vector Machine (SVM) ...")
svm_model = SVC(kernel='rbf', gamma=0.05)
svm_model.fit(X_train_flat, y_train)
y_pred_svm = svm_model.predict(X_test_flat)
svm_acc = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {svm_acc * 100:.2f}%")

# [ 5 ] Train KNN classifier
print("\nTraining K-Nearest Neighbors (KNN) ...")
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train_flat, y_train)
y_pred_knn = knn_model.predict(X_test_flat)
knn_acc = accuracy_score(y_test, y_pred_knn)
print(f"KNN Accuracy: {knn_acc * 100:.2f}%")

# [ 6 ] Compare Models
print("\nModel Performance Comparison:")
print(f"SVM Accuracy: {svm_acc * 100:.2f}%")
print(f"KNN Accuracy: {knn_acc * 100:.2f}%")

# [ 7 ] Confusion Matrix for the Best Model
best_model = "SVM" if svm_acc > knn_acc else "KNN"
y_pred_best = y_pred_svm if best_model == "SVM" else y_pred_knn

print(f"\nBest Model: {best_model}")
cm = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title(f"Confusion Matrix for {best_model}")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# [ 8 ] Classification Report
print(f"\nClassification Report for {best_model}:")
print(classification_report(y_test, y_pred_best))
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x = np.arange(0, 10.1, 0.1) # Corrected to Python numpy syntax
y = fuzz.pimf(x, 1, 4, 5, 7) # Corrected function call and parameters for skfuzzy.pimf(x, a, b, c, d)

plt.plot(x,y)
plt.xlabel('PI MF, p = (x,[1,4,5,7])')
plt.ylabel('Fuzziness')
plt.ylim([-0.05, 1.05]) # Added common ylim for better visualization
plt.title('Pi-shaped Membership Function') # Added a title for clarity
plt.grid(True) # Added a grid for clarity
plt.show()
```
